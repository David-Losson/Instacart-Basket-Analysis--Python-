```python
#create and define a price_label function

def price_label(row):
    if row ['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid range product'
    elif row ['prices'] > 15:
        return 'High range product'
    else: return np.nan
```

```python
#Apply the function to our subset
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Check the values of our subset
df['price_range'].value_counts(dropna = False)
```

```python
#Find the most expensive item in our subset
df['prices'].max()
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

4.) Create a new column, "busiest_period_of_day" that labels the busiest hours of the day.

```python
#Find the most common hours for orders
df['order_hour_of_day'].value_counts(dropna = False)
```

Observations: Because of the fact that there are 24 hours in a day, we can break the information up into three periods of 8 hours each: the most orders (9AM-5PM), the average orders (7AM-8AM & 6PM-10PM), and the least orders (11PM-6AM).

```python
#Append new labels for the busiest_period_of_day based on the order_hour_of_day column.
result = []

for value in df_ords_prods_merged ['order_hour_of_day']:
    if value in [10, 11, 14, 13, 15, 12, 16, 9] :
        result.append ('Most orders')
    elif value in [23, 6, 0, 1, 5, 2, 3, 4]:
        result.append ('Least orders')
    else:
        result.append('Average orders')
```

```python
#Add the newly-created "busiest_hours" column to the df_ords_prods_merged table
df_ords_prods_merged['busiest_periods'] = result
```

5.) Print the frequency for the new column.

```python
#Print the frequencies of each label within the busiest_hours column
print('Results of the busiest_periods column:\n', df_ords_prods_merged['busiest_hours'].value_counts())
```
